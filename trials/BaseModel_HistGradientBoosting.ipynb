# Data Preparation Pipeline

- Remove the nulls if any
- Remove the dupes
- scaling

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer
```

```python
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
from imblearn.over_sampling import SMOTE
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.metrics import f1_score
```

```python
df = raw.drop_duplicates()
```

```python
y = df['Class']
X = df.drop(columns=['Class'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
data_pipeline = Pipeline(
    [
        ('imputer', SimpleImputer()),
        ('scale', StandardScaler())
    ]
    )


data_pipeline
```

```python
data_prep = data_pipeline.fit_transform(X_train)
```

```python
smote = SMOTE()

X_smote, y_smote = smote.fit_resample(data_prep, y_train)
```

```python
models = {
    #'rfc': RandomForestClassifier(),
    'hgb': HistGradientBoostingClassifier(),
    #'KNeighbors': KNeighborsClassifier()
}
```

```python
def train_base_model(X_train, y_train, models, X_test, y_test):

  #accuracy = {}
  #precision = {}
  recall = None
  f1 = None

  for name, model in models.items():
    model.fit(X_train, y_train)
    print(name)
    y_pred = model.predict(X_test)
    #accuracy[model] = accuracy_score(y_test, y_pred)
    #precision[model] = precision_score(y_test, y_pred)

    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

  return recall, f1
```

### Training Only KNeighbors

```python
X_test_prep = data_pipeline.transform(X_test)
```

```python
train_base_model(X_smote, y_smote, models, X_test_prep, y_test)
```
